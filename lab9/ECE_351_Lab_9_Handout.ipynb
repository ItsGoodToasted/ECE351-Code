1. Plot $cos(2\pi t)$ from $0\le t \le2$s. In the same figure (separate subplot), plot the magnitude of the Fourier transform of this same signal, setting the sampling frequency $f_s=100$. Additionally, plot the phase of the Fourier transform.
```python
#Insert all necessary imports here
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack
```

```python
#Insert your function to perform the fft here
def fft_func(x, fs):

    '''
    Calculates the fast Fourier transform on the input signal.

    Inputs: x - input signal - array
            fs - sampling frequency - float
            
    Outputs: freq - output frequencies - array
             X_mag - signal magnitude - array
             X_phi - signal phase - array           
    '''
    N = len(x) #find the length of the signal
    X_fft = scipy.fftpack.fft(x) #perform the fft
    X_fft_shifted = scipy.fftpack.fftshift(X_fft) #shift zero frequency components to the center of the spectrum
    freq = np.arange(-N/2, N/2) * fs / N #compute the frequencies for the output signal. fs is the sampling frequency
    X_mag = np.abs(X_fft_shifted) / N #compute the magnitudes of the signal
    X_phi = np.angle(X_fft_shifted) #compute the phase of the signal
    
    return freq, X_mag, X_phi
```

```python
#Insert code for Part 1 Task 1 here
step = 1e-2
fs = 100
t = np.arange(0, 2 + step, step)
x = np.cos(2 * np.pi * t)

freq, mag, phi = fft_func(x, fs)

plt.figure(figsize=(8,8))
plt.suptitle('Task 1')

plt.subplot(3, 1, 1)
plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('x(t)')

freq_min, freq_max = 0, 51

plt.subplot(3, 1, 2)
plt.stem(freq, mag)
plt.xlim([freq_min, freq_max])
plt.xlabel('Frequency (Hz)')
plt.ylabel(r'X ($\omega$) Magnitude (Hz)')

plt.subplot(3, 1, 3)
plt.stem(freq, phi)
plt.xlim([freq_min, freq_max])
plt.xlabel('Frequency (Hz)')
plt.ylabel(r'X($\omega$) Phase (radians)')

plt.tight_layout()
plt.show()
```
