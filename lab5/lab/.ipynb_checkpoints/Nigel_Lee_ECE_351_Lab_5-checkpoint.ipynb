1. Plot the impulse response $h(t)$ that you found by hand in the prelab assignment from $0 \le t \le 1.2$ ms.  

```python
# Insert all necessary imports here
import scipy.signal as sig
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Insert code for your step function and the Sine Method as a function here. I will write the definition line and description as a hint.
def stepF(t):
    '''
    Step Function u(t)

    Input: t - time vector

    Output: Returns 1 for t >= 0, 0 for t < 0
    '''
    y = np.zeros(t.shape)
    for i in range(len(t)):
        if t[i] >= 0:
            y[i] = 1
        else:
            y[i] = 0
    return y
    
def sine_method(R, L, C, t):
    """
    Uses the sine method to find the impulse response of a given RLC filter,
    
    Inputs: R - Resistor value in Ohms
            L - Inductor value in Henries
            C - Capacitor value in Farads
            t - time vector to determine the response
            
    Outputs: y - impulse response calculated using the Sine Method
    """
    alpha = (-1/(2*R*C))    
    beta = 0.5*np.sqrt( (1/( (R**2) * (C**2) )) - (4/(L*C) + 0*1j ) )
    p = alpha + beta
    
    g = p*(1/(R*C))
    mag_g=np.abs(g)
    arg_g=np.angle(g)
    
    omega = np.abs(beta)

    y = (mag_g/omega)*np.exp(alpha*t)*np.sin(omega*t + arg_g)*stepF(t)

    return y
```

```python
# Insert code calling your sine method function and plotting the impulse response here

R = 1e3
L = 27e-3
C = 100e-9

step = 1e-5
t = np.arange(0, 1.2e-3 + step, step)

y = sine_method(R, L , C, t)

plt.figure(figsize=(8,4))
plt.plot(t, y)
plt.grid()
plt.ylabel("h(t)")
plt.xlabel("t [s]")
plt.title("Impulse Response of Bandpass Filter (Derived)")

plt.tight_layout()
plt.show()
```

2. Use the **scipy.signal.impulse()** function to plot the results from $0 \le t \le 1.2$ ms. 

```python
# Insert code using scipy.signal.impulse to find the impulse response here. This will be similar to the example code.
# Also, insert code plotting this response here.

R = 1e3
L = 27e-3
C = 100e-9

top = 1/(R*C)
a = 1
b = 1/(R*C)
c = 1/(L*C)

step = 1e-5
t = np.arange(0, 1.2e-3 + step, step)

num = [top,0]
den = [a,b,c]

tout, yout = sig.impulse((num, den), T = t)

plt.figure(figsize=(8,4))
plt.plot(tout, yout)
plt.grid()
plt.ylabel("h(t)")
plt.xlabel("t [s]")
plt.title("Impulse Response of Bandpass Filter (scipy)")
plt.tight_layout()
plt.show()
```

```python
# Insert code using scipy.signal.step to find the step response here. 
# Also, insert code plotting this response here.

step = 1e-5
t = np.arange(0, 1.2e-3 + step, step)

num = [top,0]
den = [a,b,c]

tout, yout = sig.step((num, den), T = t)

plt.figure(figsize=(8,4))
plt.plot(tout, yout)
plt.grid()
plt.ylabel("h(t)")
plt.xlabel("t [s]")
plt.title("Step Response of Bandpass Filter (scipy)")
plt.tight_layout()
plt.show()
```

2. Perform the final value theorem for the step response $H(s)u(s)$ in the Laplace domain.
>$FVT = \lim_{s\to\ 0 } sH(s)u(s) =\lim_{s\to\ 0 } \dfrac{\cancel{s}}{\cancel{s}}\dfrac{\dfrac{s}{RC}}{s^2+\dfrac{s}{RC}+\dfrac{1}{LC}}
= \dfrac{\dfrac{(0)}{RC}}{(0)^2+\dfrac{(0)}{RC}+\dfrac{1}{LC}} = 0$

1. Explain the result of the Final Value Theorem from **Part 2** **Task 2** in terms of the physical circuit components.  
>The zero result of the Final Value Theorem is showing the bandpass filter's behavior at extremely low frequencies. At these frequencies, the capacitor behaves as an open circuit due to small frequencies creating a large impedance, where as the inductor exhibits the inverse behavior, acting as a short due to small frequnecies creating a small impedance. This leads to current flowing only through the inductor and renders the output voltage across both elements zero. For a bandpass filter, this behavior is what allows frequencies on the low end of the spectrum to be removed from the output signal.
